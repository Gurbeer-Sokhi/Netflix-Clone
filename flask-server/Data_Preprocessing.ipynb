```python
import pandas as pd
```

```python
df = pd.read_csv("/content/movies.csv")
```

```python
df.head()
```

```python
df.drop('movieId', axis=1, inplace=True)
```

```python
df['genres'] = df['genres'].str.split('|').str[0]
```

```python
df.head()
```

```python
df['title'] = df['title'].str.replace(r'\s*\(\d+\)', '')
```

```python
df.head()
```

```python
df['title'] = df['title'].str.strip()
df['genres'] = df['genres'].str.strip()
```

```python
df.head()
```

```python
df.to_csv('recommender.csv', index=False)
```
